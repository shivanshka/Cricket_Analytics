```python
import pandas as pd
```

```python
t20_df = pd.read_csv("t20_df.csv")
```

```python
t20_df["Team_1"] = t20_df["Team_1"].apply(lambda x: "United Arab Emirates" if x.strip()=="U.A.E." else x)
t20_df["Team_2"] = t20_df["Team_2"].apply(lambda x: "United Arab Emirates" if x.strip()=="U.A.E." else x)
t20_df["Winner"] = t20_df["Winner"].apply(lambda x: "United Arab Emirates" if x.strip()=="U.A.E." else x)
```

```python
t20_df
```

```python
# For mapping Match Ids with other dataframes
match_ids = dict()
for idx, row in t20_df.iterrows():
    match_ids[f'{row["Team_1"]} vs {row["Team_2"]}'] = row["Match_Id"]
    match_ids[f'{row["Team_2"]} vs {row["Team_1"]}'] = row["Match_Id"]
```

```python
t20_df.to_csv("final_match_data.csv", index=None)
```

#### Batting data

```python
bat_df = pd.read_csv("t20_bat_summary.csv")
```

```python
bat_df["Match_Id"] = bat_df["Match"].map(match_ids)
```

```python
bat_df.head(15)
```

```python
bat_df["Dismissal"] = bat_df["Dismissal"].apply(lambda x: 'out' if x.strip() !="not out" else x.strip())
```

```python
bat_df[bat_df["SR"]=="-"]
```

```python
bat_df["SR"] = bat_df["SR"].apply(lambda x: 0 if x.strip() == "-" else x)
```

```python
bat_df[bat_df["SR"]=="-"] # All removed
```

```python
bat_df["SR"] = bat_df["SR"].astype(float)
```

```python
bat_df.info()
```

```python
bat_df.to_csv("final_bat_data.csv", index=None)
```
